# APPL Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('./processed_data/AAPL_daily_data_splits_processed.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 5))  # Set the figure size for readability

# Plotting the line graph
plt.plot(df['time_stamp'], df['close'], linestyle='-', color='b')  # Blue line graph without markers

# Setting the title and labels for the axes
plt.title('Split Adjusted APPL Stock Price vs Time')
plt.xlabel('Time')
plt.ylabel('Closing Price')

# Reverse the time on the x-axis
plt.gca().invert_xaxis()

# Adjusting the x-axis to only show labels every 100 days
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=730))  # Set major ticks to every 100 days

# Add grid lines for better readability
plt.grid(True)

# Automatically adjust subplot parameters to give the plot more padding
plt.tight_layout()

# Show the plot
plt.show()

```
